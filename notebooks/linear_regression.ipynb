```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

import sys
sys.path.append('../')

from src.linear_regression import LinearRegression
from src.optimizers import SGD, Adagrad
```

```python
theta = np.array([[0.5]])
b = np.array([-5])

def f(x: np.ndarray, theta: np.ndarray, b: np.ndarray) -> np.ndarray:
    return x @ theta.T + b + np.random.normal(0, 0.1, (x.shape[0], 1))
    # return x @ theta.T + b + np.random.normal(0, 1, (x.shape[0], 1))
```

```python
X = np.linspace(-1, 1, 100)[:, None]
Y = f(X, theta, b)
```

```python
plt.title("Data")
plt.scatter(X, Y)
```

```python
linear_regression = LinearRegression()
linear_regression.fit(X, Y, epochs=100, batch_size=1, optimizer=SGD(lr=1e-2))
```

```python
plt.plot(linear_regression.losses)
```

```python
pred_Y = linear_regression.predict(X)

plt.title("Pred")
plt.scatter(X, Y)
plt.plot(X, pred_Y)
```

```python
linear_regression.display_contour_plot()
```
