```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm

import sys
sys.path.append('../')

from src.linear_regression import LinearRegression
```

```python
theta = np.array([[2]])
b = np.array([1])

def f(x: np.ndarray, theta: np.ndarray, b: np.ndarray) -> np.ndarray:
    # return x @ theta.T + b
    return x @ theta.T + b + np.random.normal(0, 1, (x.shape[0], 1))
```

```python
X = np.linspace(-1, 1, 50)[:, None]
Y = f(X, theta, b)
```

```python
plt.title("Data")
plt.scatter(X, Y)
```

```python
linear_regression = LinearRegression()
linear_regression.fit(X, Y, epochs=1000)
```

```python
plt.plot(linear_regression.losses)
```

```python
pred_Y = linear_regression.predict(X)

plt.title("Pred")
plt.scatter(X, Y)
plt.plot(X, pred_Y)
```
